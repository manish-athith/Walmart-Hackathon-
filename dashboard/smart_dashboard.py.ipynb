```python
import streamlit as st
import pandas as pd
import plotly.express as px

# Load data
df = pd.read_csv("smart_sustain_forecast_phase2.csv")

st.set_page_config(page_title="SmartSustain Dashboard", layout="wide")

st.title("🌱 SmartSustain - Waste Intelligence Dashboard")
st.markdown("Track risk alerts, discount triggers, and sustainability actions in real-time.")

# KPIs
total_skus = len(df)
donations = (df['action'] == 'Suggest Donation').sum()
discounts = (df['action'] == 'Trigger Discount').sum()
overstock = (df['action'] == 'Monitor Overstock').sum()
no_action = (df['action'] == 'No Action').sum()

col1, col2, col3, col4, col5 = st.columns(5)
col1.metric("Total Forecasted SKUs", total_skus)
col2.metric("♻ Donation Suggestions", donations)
col3.metric("💸 Discount Triggers", discounts)
col4.metric("⚠ Overstock Monitored", overstock)
col5.metric("✅ No Action Needed", no_action)

st.markdown("---")

# 📊 Action Type Breakdown
st.subheader("📊 Action Breakdown")
action_chart = px.pie(df, names='action', title="SKU-Level Action Distribution", color_discrete_sequence=px.colors.sequential.Agsunset)
st.plotly_chart(action_chart, use_container_width=True)

# 📍 Filter by Store
store_filter = st.selectbox("🔍 Filter by Store ID", options=["All"] + sorted(df['store_id'].unique().tolist()))

if store_filter != "All":
    df = df[df['store_id'] == store_filter]

# 📦 Table View
st.subheader("📋 Inventory Risk View")
st.dataframe(df[['store_id', 'sku', 'predicted_units_sold', 'stock_on_hand', 'days_to_expiry', 'action']])

# 📊 Bar Chart by SKU
st.subheader("📊 SKU-wise Action Chart")
sku_chart = px.histogram(df, x="sku", color="action", barmode="group", title="SKU-wise Action Classification")
st.plotly_chart(sku_chart, use_container_width=True)

# Footer
st.markdown("---")
st.caption("Built with 💚 for the Walmart Hackathon | SmartSustain Project")

```
